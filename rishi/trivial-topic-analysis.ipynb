```python
import pandas as pd
import numpy as np
import pickle
```

```python
import re
import string
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer, SnowballStemmer
from nltk.stem.porter import *
```

```python
import spacy 
# python -m spacy download en_core_web_sm 
nlp = spacy.load("en_core_web_sm") 
```

```python
#not doing stemming.
def lemmatize_stemming(text):
    #stemmer = SnowballStemmer("english")
    return WordNetLemmatizer().lemmatize(text, pos='v')
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return " ".join(str(x) for x in result)
```

```python
cfp_tokens2020 = preprocess("unsupervised semi-supervised and supervised representation learning representation learning for planning and reinforcement learning metric learning and kernel learning sparse coding and dimensionality expansion hierarchical models optimization for representation learning learning representations of outputs or states theoretical issues in deep learningvisualization or interpretation of learned representationsimplementation issues parallelization software platforms hardware applications in vision audio speech natural language processing robotics neuroscience computational biology")
cfp_tokens2020 = cfp_tokens2020.split(' ')
```

```python
cfp_tokens2019 = preprocess("unsupervised semi-supervised and supervised representation learning representation learning for planning and reinforcement learning metric learning and kernel learning sparse coding and dimensionality expansion hierarchical models optimization for representation learning learning representations of outputs or states theoretical issues in deep learning visualization or interpretation of learned representations implementation issues parallelization software platforms hardware applications in vision audio speech natural language processing robotics neuroscience computational biology")
cfp_tokens2019 = cfp_tokens2019.split(' ')
```

```python
cfp_tokens2018 = preprocess("unsupervised semi-supervised and supervised representation learning representation learning for planning and reinforcement learning metric learning and kernel learning sparse coding and dimensionality expansion hierarchical models optimization for representation learning learning representations of outputs or states implementation issues parallelization software platforms hardwareapplications in vision audio speech natural language processing robotics neuroscience")
cfp_tokens2018 = cfp_tokens2018.split(' ')
```

```python
cfp_tokens2017 = preprocess("Unsupervised semi-supervised and supervised representation learning Representation learning for planning and reinforcement learning Metric learning and kernel learning Sparse coding and dimensionality expansion Hierarchical models Optimization for representation learning Learning representations of outputs or states Implementation issues parallelization software platforms hardwareApplications in vision audio speech natural language processing robotics neuroscience")
cfp_tokens2017 = cfp_tokens2017.split(' ')
```

```python
data  = pd.read_pickle(r'all_data_features_17_20.pkl')
```

```python
data['abstract'][0]
```

```python
abstract_jaccardCoef = {}
for index, abstract in zip(data['id'], data['abstract']):
    abstract = preprocess(abstract)
    if '2020' in index:
        temp = []
        for token in nlp(abstract):
            if token.pos_ == "NOUN":
                temp.append(str(token))
        try:
            abstract_jaccardCoef[index] = len(set(temp).intersection(set(cfp_tokens2020)))/len(set(temp).union(set(cfp_tokens2020)))
        except:
            continue
    elif '2019' in index:
        temp = []
        for token in nlp(abstract):
            if token.pos_ == "NOUN":
                temp.append(str(token))
        try:
            abstract_jaccardCoef[index] = len(set(temp).intersection(set(cfp_tokens2019)))/len(set(temp).union(set(cfp_tokens2019)))
        except:
            continue
    elif '2018' in index:
        temp = []
        for token in nlp(abstract):
            if token.pos_ == "NOUN":
                temp.append(str(token))
        try:
            abstract_jaccardCoef[index] = len(set(temp).intersection(set(cfp_tokens2018)))/len(set(temp).union(set(cfp_tokens2018)))
        except:
            continue
    elif '2017' in index:
        temp = []
        for token in nlp(abstract):
            if token.pos_ == "NOUN":
                temp.append(str(token))
        try:
            abstract_jaccardCoef[index] = len(set(temp).intersection(set(cfp_tokens2017)))/len(set(temp).union(set(cfp_tokens2017)))
        except:
            continue
```

```python
with open('abstract_jaccardCoef.pickle', 'wb') as handle:
    pickle.dump(abstract_jaccardCoef, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
