```python
#lets deal with 2020 papers for now
acceptAuthors = []
rejectAuthors = []
papers = papers_2020
for key in papers.keys():
    if papers[key]['DECISIONS']['decision']['content']['decision'] == 'Reject':
        for name in papers[key]['content']['authors']:
            rejectAuthors.append(name)
    else:
        for name in papers[key]['content']['authors']:
            acceptAuthors.append(name)
```

```python
data = papers_2020
accepted_author = {}
rejected_author = {}

for key in data.keys():
    if data[key]['DECISIONS']['decision']['content']['decision'] == 'Reject':
        for author in data[key]['content']['authors']:
            if author in rejected_author.keys():
                rejected_author[author]+=1
            else:
                rejected_author[author] = 1
    else:
        for author in data[key]['content']['authors']:
            if author in accepted_author.keys():
                accepted_author[author]+=1
            else:
                accepted_author[author] = 1
```

```python
accepted_author = {k: v for k, v in sorted(accepted_author.items(), key=lambda item: item[1], reverse=True)}
rejected_author = {k: v for k, v in sorted(rejected_author.items(), key=lambda item: item[1], reverse=True)}
```
